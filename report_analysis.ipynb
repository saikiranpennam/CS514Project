```python
import json
import matplotlib.pyplot as plt

def parse_bandit_json(file_path):
    """ Parse the JSON file and count issues by severity and confidence level. """
    with open(file_path, 'r') as file:
        data = json.load(file)

    # Initialize counts
    severity_counts = {'LOW': 0, 'MEDIUM': 0, 'HIGH': 0}
    confidence_counts = {'LOW': 0, 'MEDIUM': 0, 'HIGH': 0}

    # Count issues for each severity and confidence level
    for result in data['results']:
        severity = result['issue_severity']
        confidence = result['issue_confidence']
        severity_counts[severity] += 1
        confidence_counts[confidence] += 1

    return severity_counts, confidence_counts

def plot_counts(counts, title, xlabel):
    """ Plot a bar chart of the issue counts. """
    labels = counts.keys()
    values = counts.values()

    plt.bar(labels, values)
    plt.xlabel(xlabel)
    plt.ylabel('Number of Issues')
    plt.title(title)
    plt.show()

# Path to your Bandit JSON file
file_path = 'report.json'

# Parse and plot
severity_counts, confidence_counts = parse_bandit_json(file_path)
plot_counts(severity_counts, 'Bandit Vulnerability Counts by Severity Level', 'Severity Level')
plot_counts(confidence_counts, 'Bandit Vulnerability Counts by Confidence Level', 'Confidence Level')
```

```python
def parse_bandit_json(file_path):
    """ Parse the JSON file and count issues by severity and confidence level. """
    try:
        with open(file_path, 'r') as file:
            data = json.load(file)
    except json.JSONDecodeError:
        print("Error: Failed to decode JSON file. Check the file format.")
        return {}, {}, 0

    # Initialize counts
    severity_counts = {'LOW': 0, 'MEDIUM': 0, 'HIGH': 0}
    confidence_counts = {'LOW': 0, 'MEDIUM': 0, 'HIGH': 0}
    total_vulnerabilities = 0

    # Count issues for each severity and confidence level
    for result in data['results']:
        severity = result['issue_severity']
        confidence = result['issue_confidence']
        severity_counts[severity] += 1
        confidence_counts[confidence] += 1
        total_vulnerabilities += 1

    return severity_counts, confidence_counts, total_vulnerabilities

def plot_counts(counts, title, xlabel):
    """ Plot a bar chart of the issue counts. """
    labels = counts.keys()
    values = counts.values()

    plt.bar(labels, values)
    plt.xlabel(xlabel)
    plt.ylabel('Number of Issues')
    plt.title(title)
    plt.show()

# Path to your Bandit JSON file
file_path = 'report.json'

# Parse and plot
severity_counts, confidence_counts, total_vulnerabilities = parse_bandit_json(file_path)

if severity_counts and confidence_counts:
    plot_counts(severity_counts, 'Bandit Vulnerability Counts by Severity Level', 'Severity Level')
    plot_counts(confidence_counts, 'Bandit Vulnerability Counts by Confidence Level', 'Confidence Level')
    print(f"Total number of vulnerabilities found: {total_vulnerabilities}")
else:
    print("No data to display.")
```
